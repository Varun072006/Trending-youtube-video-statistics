```python
plt.figure(figsize=(8,5))
sns.countplot(x='category_id', data=df)
plt.title("Count of Videos by Category")
plt.show()
```

```python
top_channels = df.groupby("channel_title")["views"].sum().nlargest(10)
top_channels.plot(kind="bar", figsize=(10,5), title="Top 10 Channels by Total Views")
plt.show()
```

```python
from collections import Counter

tags = df['tags'].dropna().astype(str).str.split('|').sum()
top_tags = Counter(tags).most_common(10)

pd.DataFrame(top_tags, columns=["Tag", "Count"]).plot(
    x="Tag", y="Count", kind="bar", figsize=(10,5), title="Top 10 Video Tags"
)
plt.show()
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors="coerce")
plt.figure(figsize=(10,6))
plt.scatter(df['publish_time'], df['views'], alpha=0.3)
plt.title("Publish Time vs View Count")
plt.xlabel("Publish Time")
plt.ylabel("Views")
plt.show()
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors="coerce").dt.tz_localize(None)
df['trending_date'] = pd.to_datetime(df['trending_date'], errors="coerce").dt.tz_localize(None)

df['duration_days'] = (df['trending_date'] - df['publish_time']).dt.days

df['duration_days'].dropna().plot(kind="hist", bins=30, figsize=(8,5), title="Histogram of Video Duration (days)")
plt.show()
```

```python
df['month'] = df['trending_date'].dt.to_period('M')
df['month'].value_counts().sort_index().plot(kind="bar", figsize=(10,5), title="Trending Videos per Month")
plt.show()
```

```python
view_arr = df['views'].to_numpy()
like_arr = df['likes'].to_numpy()

high_eng = (df['likes'] / df['views'] > 0.1).astype(int).to_numpy()

vec_add = view_arr + like_arr

dot_prod = np.dot(view_arr, high_eng)

features = np.column_stack((view_arr, like_arr))
weights = np.array([0.6, 0.4])
weighted_score = np.dot(features, weights)

norm_views = view_arr / np.linalg.norm(view_arr)

print("Dot Product:", dot_prod)
```

```python
import sympy as sp

views, likes, dislikes = sp.symbols('views likes dislikes')
Popularity_Score = (likes*0.5) + (views*0.3) - (dislikes*0.2)
derivative_views = sp.diff(Popularity_Score, views)
derivative_views
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors="coerce").dt.tz_localize(None)

today = pd.Timestamp.now()
df['days_since_published'] = (today - df['publish_time']).dt.days
df['is_high_engagement'] = (df['likes']/df['views'] > 0.1).astype(int)
df['recent_video'] = (df['publish_time'] >= today - pd.Timedelta(days=90)).astype(int)
df['tag_count'] = df['tags'].fillna("").apply(lambda x: len(str(x).split("|")))

df['view_quartile'] = pd.qcut(df['views'], 4, labels=[1,2,3,4])
df[['title','views','view_quartile']].head()
```
